How does smoking affect insurance charges?

Does BMI significantly impact costs?

Which regions have higher average charges?

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df = pd.read_csv(r"C:\Muzna\Data Analytics\Data Analysis Projects\Medical Insurance Cost Analysis\insurance_clean.csv")
df
```

```python
sns.boxplot(x='smoker', y='charges', data=df)
plt.title("Insurance Charges: Smoker vs Non-Smoker")
plt.show()
```

```python
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df)
plt.title("BMI vs Charges")
plt.show()
```

Insights:

Smokers pay significantly higher premiums

Charges increase sharply with age + smoking

BMI has a moderate correlation with charges
