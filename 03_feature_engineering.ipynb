```python
import pandas as pd

# Read csv
df = pd.read_csv(r"C:\Muzna\Data Analytics\Data Analysis Projects\Medical Insurance Cost Analysis\insurance_clean.csv")
df
```

```python
#Create new columns

df['bmi_category'] = pd.cut(
    df['bmi'],
    bins=[0,18.5,25,30,100],
    labels=['Underweight','Normal','Overweight','Obese']
)

df['age_group'] = pd.cut(
    df['age'],
    bins=[0,25,40,60,100],
    labels=['Young','Adult','Middle-aged','Senior']
)
```
